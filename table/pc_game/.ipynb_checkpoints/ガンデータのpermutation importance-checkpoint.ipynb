```python
#Unnamedが全て欠損値、idはガンの結果に無関係なので外す。
df = df.drop(['Unnamed: 32','id'],axis=1)
df.head()
```

```python
X = df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
```

```python
X.describe()
```

```python
#各変数での箱ひげ図
plt.style.use('ggplot')#グラフ表示をみやすくする

feature_name_list = list(X.columns)
i=1
plt.figure(figsize=(10,60))
for feature_name in feature_name_list:
    plt.subplot(10, 3, i)
    df_box=df.loc[:,[feature_name,'diagnosis']]
    sns.boxplot(x='diagnosis', y=feature_name, data=df_box, hue='diagnosis')
    plt.title(feature_name)
    plt.subplots_adjust(wspace=0.6, hspace=0.5)
    i += 1
plt.show()
```

```python
#ヒストグラムの表示、青色がB、赤色がMとして表示
plt.figure(figsize=(10,30))
i = 1
for feature_name in feature_name_list:
    plt.subplot(15, 2, i)
    df_hist = df.loc[:,[feature_name,'diagnosis']].dropna(how= 'any',axis=0)
    plt.hist(df_hist.loc[df_hist['diagnosis']=='B', feature_name], color='b', alpha=0.5, label = 'B')
    plt.hist(df_hist.loc[df_hist['diagnosis']=='M',feature_name], color = 'r', alpha=0.5, label = 'M')
    plt.title(feature_name)
    plt.legend(loc='upper right')
    plt.subplots_adjust(wspace=0.3, hspace=1.0)
    plt.xlabel(feature_name)
    plt.ylabel('sum')
    i += 1
plt.show()
```

```python
#外れ値処理、whereの()内の条件を満たさないindexをNanで返す
df.loc[df['diagnosis']=='M', 'area_mean'] = df.loc[df['diagnosis']=='M', 'area_mean'].where(df.loc[df['diagnosis']=='M', 'area_mean'] < 2200)
df.loc[df['diagnosis']=='B', 'smoothness_mean'] = df.loc[df['diagnosis']=='B', 'smoothness_mean'].where(df.loc[df['diagnosis']=='B', 'smoothness_mean'] < 0.15)
df.loc[df['diagnosis']=='M', 'symmetry_mean'] = df.loc[df['diagnosis']=='M', 'symmetry_mean'].where(df.loc[df['diagnosis']=='M', 'symmetry_mean'] < 0.25)
df.loc[df['diagnosis']=='M', 'radius_se'] = df.loc[df['diagnosis']=='M', 'radius_se'].where(df.loc[df['diagnosis']=='M', 'radius_se'] < 2)
df.loc[df['diagnosis']=='B', 'texture_se'] = df.loc[df['diagnosis']=='B', 'texture_se'].where(df.loc[df['diagnosis']=='B', 'texture_se'] < 3.3)
df.loc[df['diagnosis']=='M', 'perimeter_se'] = df.loc[df['diagnosis']=='M', 'perimeter_se'].where(df.loc[df['diagnosis']=='M', 'perimeter_se'] < 15)
df.loc[df['diagnosis']=='M', 'area_se'] = df.loc[df['diagnosis']=='M', 'area_se'].where(df.loc[df['diagnosis']=='M', 'area_se'] < 400)
df.loc[df['diagnosis']=='B', 'concave points_se'] = df.loc[df['diagnosis']=='B', 'concave points_se'].where(df.loc[df['diagnosis']=='B', 'concave points_se'] < 0.04)
df.loc[df['diagnosis']=='M', 'fractal_dimension_worst'] = df.loc[df['diagnosis']=='M', 'fractal_dimension_worst'].where(df.loc[df['diagnosis']=='M', 'fractal_dimension_worst'] < 0.15)
```

```python
df.isnull().sum()
```

```python
#欠損のある列を削除
df = df.dropna()
```

```python
df.shape
```

```python
#目的変数のマッピング
target_mapping = {'B':0, 'M':1}
df['diagnosis']  =df['diagnosis'].map(target_mapping)
```

```python
df.head()
```

```python
X = df.drop(['diagnosis'],axis=1)
y = df['diagnosis']
```

```python
random_state = 0
#今回はsvmのモデルを扱うので、標準化をして数字の大きさによる変数の重要度の依存性を無くす
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_std = scaler.transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_std, y, test_size=0.3, random_state=random_state)
```

```python
param_grid = {'C': [1, 3, 5,7,10],
              'gamma': [0.001, 0.01, 0.1]}
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
kf_5 = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)
#カーネルSVMで回帰モデル作成
from sklearn.svm import SVC
gs_svc = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=kf_5)

gs_svc.fit(X_train, y_train)
print(gs_svc.best_params_)
print(gs_svc.best_score_)
print(gs_svc.score(X_test, y_test))
```

```python
#grid_searchで得た一番いいparamのモデルを取得
best_model = gs_svc.best_estimator_

#混同行列の作成
from sklearn.metrics import confusion_matrix

cm_train = confusion_matrix(y_train, best_model.predict(X_train))
cm_test = confusion_matrix(y_test, best_model.predict(X_test))

def plot_cm(confmat):
    fig, ax = plt.subplots(figsize=(5, 5))
    ax.matshow(confmat, cmap=plt.cm.Blues, alpha=0.3)
    for i in range(confmat.shape[0]):
        for j in range(confmat.shape[1]):
            ax.text(x=j, y=i, s=confmat[i, j], va='center', ha='center')
    plt.title('predicted label')
    plt.ylabel('true label')
    plt.tight_layout() 
    plt.show()
```

```python
plot_cm(cm_train)
plot_cm(cm_test)
```

```python
import eli5
from eli5.sklearn import PermutationImportance
#SVMのモデル内での変数の重要度を得るためのライブラリを取得
```

```python
#作成したモデルに検証データをかけたインスタンスを作成
perm = PermutationImportance(best_model,random_state=random_state).fit(X_test, y_test)

#重要度の高い変数を上から表示
#explaim_weights_dfにするとdataframeで返してくれる
#eli5.show_weights(インスタンス(必ず設定),feature_names=カラム名(リスト型))の引数設定、他引数アリ、公式ドキュメント参照
#https://eli5.readthedocs.io/en/latest/autodocs/eli5.html ←公式document
#https://bunseki-train.com/permutaion_importance_and_eli5_2/ ←参考リンク
eli5.show_weights(perm, feature_names=X.columns.tolist())
```

```python
#show_predictionにすると、一つのデータをモデルにかけたときの重要度が算出される。
#今回はデータ数が550件ほどあるので、X_testのデータ数分だけそれぞれの変数の重要度がわかる
#eli5では、カーネルSVMは扱っておらず、linearSVMのみ扱っている
#下記でランダムフォレストで再挑戦してます
eli5.show_prediction(best_model, doc=X_test[0], feature_names= X.columns.tolist(), show_feature_values=True)
```

```python
#重要度の上位9つを用いて、再度SVMを作成
X_perm = df[['texture_worst','compactness_se','smoothness_worst','radius_worst','concave points_worst',
            'area_mean','texture_se','texture_mean','symmetry_worst']]
y = df['diagnosis']
```

```python
X_perm.head()
```

```python
#標準化
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X_perm)
X_perm_std = scaler.transform(X_perm)
```

```python
from sklearn.model_selection import train_test_split
X_perm_train,X_perm_test,y_perm_train,y_perm_test = train_test_split(X_perm_std, y, test_size=0.3, random_state=random_state)
```

```python
param_grid = {'C': [5,7,9,11],
              'gamma': [0.01, 0.1,1]}

kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

gs_svc_perm = GridSearchCV(SVC(kernel='rbf'), param_grid, cv=kf)

gs_svc_perm.fit(X_perm_train, y_perm_train)
print(gs_svc_perm.best_params_)
print(gs_svc_perm.best_score_)
print(gs_svc_perm.score(X_perm_test, y_perm_test))
```

```python
best_model_perm = gs_svc_perm.best_estimator_

cm_train_perm = confusion_matrix(y_train, best_model_perm.predict(X_perm_train))
cm_test_perm = confusion_matrix(y_test, best_model_perm.predict(X_perm_test))
```

```python
plot_cm(cm_train_perm)
plot_cm(cm_test_perm)
```

```python
print('結果的に、どの変数も寄与率が小さいため、上位の変数だけを使うより全ての変数をモデルに入れた方が制度が良くなりそう。またweightの標準偏差の方が平均値より大きい特徴量は、もしかしたらモデルに入れることで悪影響を及ぼす可能性がありそう')
```

```python
#線形モデルだとshow_predictionが算出できないので、ランダムフォレストで挑戦、ランダムフォレストはeli5でなくても重要度がわかるが今回はeli5でやる
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

param_grid = {'n_estimators':[1000], 'max_depth':[3,5,7,9]}
grid_search = GridSearchCV(RandomForestClassifier(),param_grid,cv=5,scoring='roc_auc').fit(X_train,y_train)


print('trainデータのスコア{}'.format(grid_search.score(X_train, y_train)))
print('テストデータのスコア{}'.format(grid_search.score(X_test, y_test)))
print("best_params:{}".format(grid_search.best_params_))

```

```python
#show_weightsでモデル全体の変数の重要度を算出

best_model_rfc = grid_search.best_estimator_
#作成したモデルに検証データをかけたインスタンスを作成
perm_rfc = PermutationImportance(best_model_rfc,random_state=random_state).fit(X_test, y_test)

#重要度の高い変数を上から表示
#explaim_weights_dfにするとdataframeで返してくれる
#eli5.show_weights(インスタンス(必ず設定),feature_names=カラム名(リスト型))の引数設定、他引数アリ、公式ドキュメント参照
#https://eli5.readthedocs.io/en/latest/autodocs/eli5.html ←公式document
#https://bunseki-train.com/permutaion_importance_and_eli5_2/ ←参考リンク
eli5.explain_weights(perm_rfc, feature_names=X.columns.tolist())
```

```python
#ランダムフォレストでshow_predictionを算出
#show_weightsとことなり、引数のインスタンスはpermでないことに注意
#show_prediction(インスタンス(必ず設定),doc=(どの行のデータでの変数重要度をだしたいか、必ず設定）,feature_names = カラム名(リスト型))
#show_feature_values=Trueだと機能の寄与と共に機能の値だしてくれる。
#https://eli5.readthedocs.io/en/latest/autodocs/eli5.html ←引数の参考リンク
eli5.show_prediction(best_model_rfc, doc=X_test[160], feature_names= X.columns.tolist(), show_feature_values=True)
```
